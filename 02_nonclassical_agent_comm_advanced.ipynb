## The logics in this notebook — definitions and references

### 1. Computation Tree Logic (CTL)
**Definition.** A branching-time temporal logic interpreted over Kripke transition systems. Combines path quantifiers `A` ("on all paths"), `E` ("on some path") with state operators `X`, `F`, `G`, `U`. `AF φ` says "on every path, eventually `φ`".
**Reference.** Clarke, E. M. & Emerson, E. A. (1981). *Design and synthesis of synchronization skeletons using branching-time temporal logic*. Logic of Programs Workshop, LNCS 131, 52–71.

### 2. Alternating-time Temporal Logic (ATL)
**Definition.** Generalises CTL by replacing path quantifiers with *coalition* quantifiers `<<C>>φ` = "the coalition `C` has a joint strategy to enforce `φ`". Interpreted over concurrent game structures.
**Reference.** Alur, R., Henzinger, T. A. & Kupferman, O. (2002). *Alternating-time temporal logic*. Journal of the ACM 49(5), 672–713.

### 3. Dynamic Epistemic Logic (DEL)
**Definition.** Extends epistemic logic with *events* described by **action models** `(E, R_i, pre)`. The product update `M ⊗ A` produces a new epistemic model in which agents' uncertainty about the event is composed with their uncertainty about the world. Captures private, semi-private, and public updates uniformly.
**Reference.** Baltag, A., Moss, L. S. & Solecki, S. (1998). *The logic of public announcements, common knowledge, and private suspicions*. TARK '98, 43–56. Textbook: van Ditmarsch, van der Hoek & Kooi, *Dynamic Epistemic Logic* (Springer 2007).

### 4. Public Announcement Logic & common knowledge
**Definition.** PAL extends epistemic logic with `[!φ]ψ` = "after announcing `φ`, `ψ` holds". Common knowledge `Cφ` of a group `G` is the greatest fixed point of "everyone in `G` knows that `φ` and that `Cφ`"; equivalently, `φ` holds at every world reachable by the transitive closure of `⋃_{i∈G} R_i`.
**Reference.** Plaza, J. (1989). *Logics of public communications*. ISMIS '89, 201–216. Reprinted in Synthese 158, 165–179 (2007).

### 5. AGM belief revision
**Definition.** A theory of how a rational agent should change its belief set when receiving new information. The AGM postulates characterise three operations: expansion, contraction, and revision. A common semantics uses *plausibility rankings* on possible worlds; revision keeps the most plausible worlds compatible with the new fact.
**Reference.** Alchourrón, C. E., Gärdenfors, P. & Makinson, D. (1985). *On the logic of theory change: partial meet contraction and revision functions*. Journal of Symbolic Logic 50(2), 510–530.

### 6. Possibilistic logic
**Definition.** Each formula in the knowledge base carries a *necessity weight* `N(φ) ∈ [0,1]`. Inference rules propagate weights by taking the minimum along resolution chains. The *inconsistency level* of a base is the largest weight `α` such that the `α`-cut is inconsistent.
**Reference.** Dubois, D., Lang, J. & Prade, H. (1994). *Possibilistic logic*. In *Handbook of Logic in AI and Logic Programming*, Vol. 3, OUP, 439–513.

### 7. Subjective logic
**Definition.** A calculus of *opinions* `ω = (b, d, u, a)` over a binary frame, where `b + d + u = 1` and `a` is the prior base rate. Operators include cumulative fusion (independent sources), averaging fusion (dependent sources), and discounting (trust composition).
**Reference.** Jøsang, A. (2001). *A logic for uncertain probabilities*. International Journal of Uncertainty, Fuzziness and Knowledge-Based Systems 9(3), 279–311. Book: Jøsang, *Subjective Logic* (Springer 2016).

### 8. Dempster–Shafer theory of evidence
**Definition.** A generalisation of probability in which belief mass `m: 2^Ω → [0,1]` is distributed over *subsets* of a frame of discernment `Ω`, with `m(∅)=0` and `Σ m(A)=1`. Dempster's rule combines two independent mass functions and renormalises by the conflict mass `K`.
**Reference.** Dempster, A. P. (1967). *Upper and lower probabilities induced by a multivalued mapping*. Annals of Mathematical Statistics 38, 325–339. Shafer, G. (1976). *A Mathematical Theory of Evidence*. Princeton.

### 9. Bilattice logic (Belnap / Ginsberg)
**Definition.** A bilattice is a set with two distinct lattice orders: a *truth* order `≤_t` and an *information / knowledge* order `≤_k`. Belnap's `FOUR = {N, T, F, B}` is the simplest example. Information from independent sources is fused via `≤_k`-join; truth-functional reasoning uses `≤_t`-meet/join.
**Reference.** Belnap, N. D. (1977). *A useful four-valued logic*. In Dunn & Epstein (eds.), *Modern Uses of Multiple-Valued Logic*. Reidel, 5–37. Ginsberg, M. L. (1988). *Multivalued logics: a uniform approach to inference in artificial intelligence*. Computational Intelligence 4(3), 265–316.

### 10. Free logic
**Definition.** A first-order logic in which singular terms may fail to denote. The existence predicate `E!t` is first-class. *Negative* free logic makes any atomic formula with a non-denoting term false; *positive* free logic allows extra-domain valuations; *neutral* free logic introduces a third truth value.
**Reference.** Lambert, K. (1967). *Free logic and the concept of existence*. Notre Dame Journal of Formal Logic 8, 133–144. Survey: Bencivenga, E. (1986). *Free logics*. Handbook of Philosophical Logic III, 373–426.

### 11. Standard Deontic Logic (SDL)
**Definition.** A normal modal logic (KD) whose box operator is read `Oφ` = "`φ` is obligatory". The accessibility relation links a world to its *deontically ideal* alternatives. Permission `Pφ ≡ ¬O¬φ`, prohibition `Fφ ≡ O¬φ`. SDL famously suffers from the Ross and Chisholm paradoxes, motivating richer deontic systems.
**Reference.** von Wright, G. H. (1951). *Deontic logic*. Mind 60(237), 1–15.

### 12. Quantum logic
**Definition.** Propositions about a quantum system correspond to closed subspaces of a Hilbert space. Conjunction is intersection of subspaces, disjunction is the closed linear span, negation is orthogonal complement. The lattice is **orthomodular** but not distributive: `a ∧ (b ∨ c) ≠ (a ∧ b) ∨ (a ∧ c)` in general.
**Reference.** Birkhoff, G. & von Neumann, J. (1936). *The logic of quantum mechanics*. Annals of Mathematics 37(4), 823–843.

### 13. Linear logic
**Definition.** A *substructural* logic that drops the structural rules of weakening and contraction, so each premise is consumed exactly once. Multiplicative connectives include tensor `⊗` ("both, jointly") and linear implication `⊸` ("consume the left, produce the right"). The exponential `!φ` reintroduces classical reuse.
**Reference.** Girard, J.-Y. (1987). *Linear logic*. Theoretical Computer Science 50(1), 1–101.

### 14. Independence-Friendly (IF) logic
**Definition.** Extends first-order logic by allowing slashed quantifiers `∃y/∀x` meaning "choose `y` independently of `x`". Game-theoretic semantics: the verifier and falsifier play a game with imperfect information; an IF formula is true iff the verifier has a winning strategy that respects the independence constraints.
**Reference.** Hintikka, J. & Sandu, G. (1989). *Informational independence as a semantical phenomenon*. Logic, Methodology and Philosophy of Science VIII, 571–589. Book: Hintikka, *The Principles of Mathematics Revisited* (CUP 1996).

## 1. Computation Tree Logic (CTL)

Branching-time temporal logic over a Kripke transition system. Path quantifiers `A` (all paths), `E` (some path) combined with `X`, `F`, `G`, `U`.

**Scenario.** A two-agent handshake protocol with non-deterministic choices. We ask: from the initial state, is it true that on **every** path, the protocol **eventually** reaches `done`?

```python
@dataclass
class TS:
    states: Set[str]
    trans: Dict[str, Set[str]]
    label: Dict[str, Set[str]]   # state -> atoms true there

def ctl_sat(ts: TS, f) -> Set[str]:
    op = f[0]
    if op == "atom":
        return {s for s in ts.states if f[1] in ts.label.get(s, set())}
    if op == "not":
        return ts.states - ctl_sat(ts, f[1])
    if op == "and":
        return ctl_sat(ts, f[1]) & ctl_sat(ts, f[2])
    if op == "or":
        return ctl_sat(ts, f[1]) | ctl_sat(ts, f[2])
    if op == "EX":
        sub = ctl_sat(ts, f[1])
        return {s for s in ts.states if ts.trans.get(s, set()) & sub}
    if op == "AX":
        sub = ctl_sat(ts, f[1])
        return {s for s in ts.states if ts.trans.get(s, set()) and ts.trans[s] <= sub}
    if op == "EF":      # least fixed point
        return ctl_sat(ts, ("EU", ("atom", "_true"), f[1]))
    if op == "AF":
        # AF φ = least Y. φ ∨ AX Y
        Y = ctl_sat(ts, f[1])
        while True:
            new = Y | {s for s in ts.states if ts.trans.get(s, set()) and ts.trans[s] <= Y}
            if new == Y: return Y
            Y = new
    if op == "EG":
        # greatest fixed point: φ ∧ EX Y
        Y = ctl_sat(ts, f[1])
        while True:
            new = Y & {s for s in ts.states if ts.trans.get(s, set()) & Y}
            if new == Y: return Y
            Y = new
    if op == "EU":
        phi, psi = ctl_sat(ts, f[1]), ctl_sat(ts, f[2])
        if f[1] == ("atom","_true"): phi = ts.states
        Y = set(psi)
        while True:
            new = Y | {s for s in phi if ts.trans.get(s, set()) & Y}
            if new == Y: return Y
            Y = new
    raise ValueError(op)

ts = TS(
    states={"s0","s1","s2","s3","sErr"},
    trans={"s0":{"s1","sErr"}, "s1":{"s2"}, "s2":{"s3"}, "s3":{"s3"}, "sErr":{"sErr"}},
    label={"s3":{"done"}, "sErr":{"err"}},
)
print("AF done holds at:", ctl_sat(ts, ("AF", ("atom","done"))))
print("EF done holds at:", ctl_sat(ts, ("EF", ("atom","done"))))
print("\nA non-deterministic branch into sErr breaks AF done at s0 — exactly the kind")
print("of liveness bug that linear-time properties on a single trace can mask.")
```

## 2. Alternating-time Temporal Logic (ATL)

ATL replaces CTL's path quantifiers with **coalition** quantifiers `<<C>>φ` = "coalition C has a joint strategy to enforce φ". Perfect for asking *which group of agents can guarantee an outcome*.

**Scenario.** A 2-player concurrent game. Agent 1 wants to reach `goal`; agent 2 tries to block. We compute `<<{1}>> F goal` and `<<{1,2}>> F goal`.

```python
@dataclass
class CGS:
    """Concurrent game structure."""
    states: Set[str]
    actions: Dict[int, List[str]]              # agent -> available actions (uniform)
    trans: Dict[Tuple[str, Tuple[str, ...]], str]  # (state, joint action) -> next
    label: Dict[str, Set[str]]
    agents: List[int]

def atl_can_force(cgs: CGS, coalition: Set[int], target: Set[str]) -> Set[str]:
    """States from which `coalition` has a memoryless strategy to eventually reach target."""
    Y = set(target)
    while True:
        new = set(Y)
        for s in cgs.states - Y:
            coal_acts  = list(product(*[cgs.actions[a] for a in coalition]))
            other_idx  = [a for a in cgs.agents if a not in coalition]
            other_acts = list(product(*[cgs.actions[a] for a in other_idx]))
            for ca in coal_acts:
                ok = True
                for oa in other_acts:
                    joint = []
                    ci, oi = 0, 0
                    for a in cgs.agents:
                        if a in coalition: joint.append(ca[ci]); ci += 1
                        else:               joint.append(oa[oi]); oi += 1
                    nxt = cgs.trans.get((s, tuple(joint)))
                    if nxt is None or nxt not in Y:
                        ok = False; break
                if ok:
                    new.add(s); break
        if new == Y: return Y
        Y = new

# Toy game: two doors. Agent 1 picks L/R, agent 2 picks L/R. Match -> goal, mismatch -> retry.
cgs = CGS(
    states={"play","goal"},
    actions={1:["L","R"], 2:["L","R"]},
    trans={
        ("play",("L","L")):"goal", ("play",("R","R")):"goal",
        ("play",("L","R")):"play", ("play",("R","L")):"play",
        ("goal",("L","L")):"goal", ("goal",("L","R")):"goal",
        ("goal",("R","L")):"goal", ("goal",("R","R")):"goal",
    },
    label={"goal":{"goal"}},
    agents=[1,2],
)
target = {"goal"}
print("<<{1}>> F goal      =", atl_can_force(cgs, {1}, target))
print("<<{1,2}>> F goal    =", atl_can_force(cgs, {1,2}, target))
print("\nAgent 1 alone cannot force matching — needs cooperation. ATL makes this crisp.")
```

## 3. Dynamic Epistemic Logic (DEL) with action models

Extends epistemic logic with **events** that may be observed differently by different agents. Each event `e` has a precondition; agent `i`'s relation on events captures what `i` cannot distinguish.

**Scenario.** Three agents A, B, C. A privately learns `p` while B and C only see that *something* happened. We compute the product update.

```python
@dataclass
class EpiM:
    worlds: Set[str]
    R: Dict[str, Set[Tuple[str,str]]]
    V: Dict[str, Set[str]]

    def holds(self, f, w):
        if f[0]=="atom": return w in self.V.get(f[1], set())
        if f[0]=="not":  return not self.holds(f[1], w)
        if f[0]=="and":  return self.holds(f[1],w) and self.holds(f[2],w)
        if f[0]=="K":
            return all(self.holds(f[2], v) for (u,v) in self.R[f[1]] if u==w)
        raise ValueError(f)

@dataclass
class ActionModel:
    events: Set[str]
    pre: Dict[str, tuple]
    R: Dict[str, Set[Tuple[str,str]]]

def product_update(M: EpiM, A: ActionModel) -> EpiM:
    new_worlds = {(w,e) for w in M.worlds for e in A.events if M.holds(A.pre[e], w)}
    name = lambda we: f"{we[0]}|{we[1]}"
    NW = {name(we) for we in new_worlds}
    NR = {}
    for ag in M.R:
        rel = set()
        for (w,e) in new_worlds:
            for (w2,e2) in new_worlds:
                if (w,w2) in M.R[ag] and (e,e2) in A.R[ag]:
                    rel.add((name((w,e)), name((w2,e2))))
        NR[ag] = rel
    NV = {p: {name(we) for we in new_worlds if we[0] in M.V.get(p,set())} for p in M.V}
    return EpiM(NW, NR, NV)

# Initial model: two worlds, p true in one. Nobody knows which.
M = EpiM(
    worlds={"wp","wn"},
    R={ag: {("wp","wp"),("wn","wn"),("wp","wn"),("wn","wp")} for ag in ("A","B","C")},
    V={"p":{"wp"}},
)
# Action model: A privately learns whether p; B and C see only "something happened".
A = ActionModel(
    events={"learnP","learnNotP"},
    pre={"learnP":("atom","p"), "learnNotP":("not",("atom","p"))},
    R={
        "A": {("learnP","learnP"),("learnNotP","learnNotP")},
        "B": {("learnP","learnP"),("learnP","learnNotP"),("learnNotP","learnP"),("learnNotP","learnNotP")},
        "C": {("learnP","learnP"),("learnP","learnNotP"),("learnNotP","learnP"),("learnNotP","learnNotP")},
    },
)
M2 = product_update(M, A)
actual = "wp|learnP"
print("After private learning:")
print("  A knows p?", M2.holds(("K","A",("atom","p")), actual))
print("  B knows p?", M2.holds(("K","B",("atom","p")), actual))
print("  B knows that A knows whether p?",
      M2.holds(("K","B",("K","A",("atom","p"))), actual)
      or M2.holds(("K","B",("K","A",("not",("atom","p")))), actual))
```

## 4. Common knowledge after public announcement

Common knowledge `Cφ` = everyone knows, everyone knows that everyone knows, ad infinitum. Computed as the transitive closure of the union of agents' relations.

```python
def transitive_closure(rel):
    closure = set(rel)
    while True:
        new = closure | {(a,c) for (a,b) in closure for (b2,c) in closure if b==b2}
        if new == closure: return closure
        closure = new

def common_knowledge(M: EpiM, atom: str, w: str, agents: List[str]) -> bool:
    union = set().union(*[M.R[a] for a in agents])
    cl = transitive_closure(union)
    return all(v in M.V.get(atom,set()) for (u,v) in cl if u==w) and w in M.V.get(atom,set())

# Reusing the simple two-world model from sec 3 but after a *public* announcement of p
M = EpiM(
    worlds={"wp","wn"},
    R={ag: {("wp","wp"),("wn","wn"),("wp","wn"),("wn","wp")} for ag in ("A","B","C")},
    V={"p":{"wp"}},
)
# Public announce p: keep only worlds where p, drop other accessibilities
kept = {"wp"}
Mpub = EpiM(
    worlds=kept,
    R={ag: {("wp","wp")} for ag in ("A","B","C")},
    V={"p":{"wp"}},
)
print("Common knowledge of p after public announcement?",
      common_knowledge(Mpub, "p", "wp", ["A","B","C"]))
```

## 5. Belief revision (AGM-style, ranking-based)

An agent stores a **plausibility ranking** over possible worlds. New information triggers revision: keep the most plausible worlds compatible with the new fact.

**Scenario.** Agent receives a report it had ranked as implausible. After revision, the previously-second-most-plausible worlds become the new belief set.

```python
@dataclass
class Ranking:
    # rank 0 = most plausible
    worlds: Dict[str, int]              # world id -> rank
    valuation: Dict[str, Set[str]]      # atom -> worlds where true

    def belief_set(self):
        m = min(self.worlds.values())
        return {w for w,r in self.worlds.items() if r == m}

    def revise(self, atom):
        compat = self.valuation.get(atom, set())
        if not compat: return self
        m = min(self.worlds[w] for w in compat)
        new_ranks = {}
        for w, r in self.worlds.items():
            if w in compat: new_ranks[w] = r - m       # promote compatible worlds
            else:           new_ranks[w] = r - m + 1   # push incompatible down
        return Ranking(new_ranks, self.valuation)

R = Ranking(
    worlds={"w_p":0, "w_q":1, "w_pq":2, "w_none":3},
    valuation={"p":{"w_p","w_pq"}, "q":{"w_q","w_pq"}},
)
print("Initial belief set:", R.belief_set())
R2 = R.revise("q")
print("After revising with q:", R2.belief_set())
print("\nClassical KB updates either accept or contradict; AGM picks the most plausible repair.")
```

## 6. Possibilistic logic

Each formula in the KB carries a **necessity weight** in `[0,1]`. Inferences inherit the minimum weight along the chain. Useful when agents communicate facts of varying reliability.

```python
from itertools import combinations
from collections import defaultdict

def neg(lit):
    return (lit[0], not lit[1])

def possibilistic_resolve_optimized(kb, max_clauses=10000):
    """Optimized: uses hashing, subsumption, and level-by-level generation."""
    # clause -> best necessity (higher is better certainty)
    best = {}
    for c, n in kb:
        key = frozenset(c)
        if key not in best or n > best[key]:
            best[key] = n

    changed = True
    rounds = 0
    while changed and rounds < 100:
        changed = False
        rounds += 1
        new_resolvents = defaultdict(lambda: 0)

        clauses_list = list(best.items())
        for (c1, n1), (c2, n2) in combinations(clauses_list, 2):
            # Find complementary literals
            lits1 = frozenset(c1)
            lits2 = frozenset(c2)
            for l in c1:
                nl = neg(l)
                if nl in lits2:
                    # Compute resolvent
                    res = (lits1 - {l}) | (lits2 - {nl})
                    nn = min(n1, n2)
                    key = frozenset(res)
                    # Only keep if this is a genuinely new / better derivation
                    if key not in best or nn > new_resolvents[key]:
                        new_resolvents[key] = nn

        # Add new resolvents, prune by subsumption
        for res_key, nn in new_resolvents.items():
            if res_key not in best or nn > best[res_key]:
                # Check subsumption: does any existing clause subsume res_key?
                subsumed = False
                for existing_key, en in best.items():
                    if existing_key.issubset(res_key) and en >= nn:
                        subsumed = True
                        break
                if not subsumed and len(best) < max_clauses:
                    best[res_key] = nn
                    changed = True

    return [(set(c), n) for c, n in best.items()]

kb = [
    ({("bird", False), ("flies", True)}, 0.8),
    ({("bird", True)},                    0.9),
    ({("penguin", False), ("flies", False)}, 1.0),
]
closed = possibilistic_resolve_optimized(kb)
for c, n in sorted(closed, key=lambda x: -x[1]):
    print(f"  necessity {n:.1f}: {sorted(c)}")
```

## 7. Subjective logic — opinions with uncertainty

An opinion is a tuple `(b, d, u, a)`: belief, disbelief, uncertainty (`b+d+u=1`) and base rate `a`. Supports cumulative fusion of independent agents' opinions.

```python
@dataclass
class Opinion:
    b: float; d: float; u: float; a: float = 0.5
    def projected(self): return self.b + self.a * self.u

def cumulative_fusion(o1: Opinion, o2: Opinion) -> Opinion:
    if o1.u == 0 and o2.u == 0:
        return Opinion((o1.b+o2.b)/2, (o1.d+o2.d)/2, 0.0, o1.a)
    denom = o1.u + o2.u - o1.u*o2.u
    b = (o1.b*o2.u + o2.b*o1.u) / denom
    d = (o1.d*o2.u + o2.d*o1.u) / denom
    u = (o1.u*o2.u) / denom
    return Opinion(b,d,u,o1.a)

agentA = Opinion(b=0.7, d=0.1, u=0.2)
agentB = Opinion(b=0.4, d=0.4, u=0.2)
fused  = cumulative_fusion(agentA, agentB)
print("A:", agentA, "-> projected", agentA.projected())
print("B:", agentB, "-> projected", agentB.projected())
print("Fused:", fused, "-> projected", fused.projected())
```

## 8. Dempster–Shafer evidence combination

Belief mass distributed over **subsets** of a frame of discernment. Dempster's rule fuses two independent evidence sources, normalising for conflict.

```python
def dempster(m1: Dict[FrozenSet[str], float], m2: Dict[FrozenSet[str], float]):
    combined = {}
    K = 0.0
    for A, pA in m1.items():
        for B, pB in m2.items():
            inter = A & B
            if not inter:
                K += pA*pB
            else:
                combined[inter] = combined.get(inter, 0) + pA*pB
    if K >= 1.0: raise ValueError("total conflict")
    return {s: v/(1-K) for s,v in combined.items()}

F = lambda *xs: frozenset(xs)
# Frame: which suspect committed the act?
m_witness1 = {F("alice"):0.6, F("alice","bob"):0.3, F("alice","bob","carol"):0.1}
m_witness2 = {F("bob"):0.5,   F("alice","bob"):0.4, F("alice","bob","carol"):0.1}
fused = dempster(m_witness1, m_witness2)
for s,p in sorted(fused.items(), key=lambda x:-x[1]):
    print(f"  {set(s)}: {p:.3f}")
```

## 9. Bilattice (Belnap 4-valued, Ginsberg form)

Two orders on the same 4 values `{N, T, F, B}`: a *truth* order (`F ≤_t N,B ≤_t T`) and a *knowledge* order (`N ≤_k T,F ≤_k B`). Agents fuse information along ≤_k and reason along ≤_t.

```python
N,T,F,B = "N","T","F","B"
# Truth lattice meets/joins
_t_order = {F:0, N:1, B:1, T:2}
_k_order = {N:0, T:1, F:1, B:2}
def t_meet(a,b):
    cands = [v for v in (N,T,F,B) if _t_order[v]<=_t_order[a] and _t_order[v]<=_t_order[b]]
    return max(cands, key=lambda v:_t_order[v])
def t_join(a,b):
    cands = [v for v in (N,T,F,B) if _t_order[v]>=_t_order[a] and _t_order[v]>=_t_order[b]]
    return min(cands, key=lambda v:_t_order[v])
def k_join(a,b):    # consensus / information fusion
    cands = [v for v in (N,T,F,B) if _k_order[v]>=_k_order[a] and _k_order[v]>=_k_order[b]]
    return min(cands, key=lambda v:_k_order[v])

a_says = T; b_says = F
print(f"Knowledge fusion of T and F → {k_join(a_says,b_says)}  (B = both, conflict made explicit)")
print(f"Truth join     of T and N → {t_join(T,N)}")
print(f"Truth meet     of T and B → {t_meet(T,B)}")
```

## 10. Free logic

Allows terms that fail to denote. Predicates over non-existing objects evaluate to a designated `*` (or false, depending on the variant). Useful for agents talking about hypothetical or yet-uncreated entities.

```python
@dataclass
class FreeModel:
    domain: Set[str]            # things that exist
    outer:  Set[str]            # things that can be named but don't exist
    interp: Dict[str, Set[str]] # predicate -> tuple-set

    def E(self, t):             # existence predicate
        return t in self.domain

    def eval_pred(self, P, t):
        if not self.E(t): return "*"        # negative free logic style
        return t in self.interp.get(P,set())

M = FreeModel(
    domain={"alice","bob"},
    outer={"merlin"},                       # legendary, can be referenced
    interp={"wizard":{"merlin"}, "agent":{"alice","bob"}},
)
for t in ("alice","merlin"):
    print(f"  exists({t})={M.E(t)}, agent({t})={M.eval_pred('agent',t)}, wizard({t})={M.eval_pred('wizard',t)}")
print("\nClassical FOL forces every name to denote — agents discussing hypothetical referents can't.")
```

## 11. Deontic logic (Standard Deontic Logic, SDL)

`Oφ` = obligatory, `Pφ` = permitted, `Fφ` = forbidden. Modelled as a Kripke frame whose accessible worlds are the *deontically ideal* ones. Useful for protocols with norms.

**Scenario.** Agent X is obligated to acknowledge any request. We check whether a trace where X stays silent satisfies its obligations.

```python
@dataclass
class DeonticModel:
    worlds: Set[str]
    ideal: Dict[str, Set[str]]   # world -> ideal alternatives
    label: Dict[str, Set[str]]

    def holds(self, f, w):
        if f[0]=="atom": return f[1] in self.label.get(w,set())
        if f[0]=="not":  return not self.holds(f[1],w)
        if f[0]=="and":  return self.holds(f[1],w) and self.holds(f[2],w)
        if f[0]=="O":    return all(self.holds(f[1],v) for v in self.ideal[w])
        if f[0]=="P":    return any(self.holds(f[1],v) for v in self.ideal[w])
        raise ValueError(f)

M = DeonticModel(
    worlds={"silent","ack","reject"},
    ideal={"silent":{"ack"}, "ack":{"ack"}, "reject":{"ack"}},
    label={"ack":{"ack"}, "reject":{"reject"}},
)
print("At 'silent': O(ack)?", M.holds(("O",("atom","ack")),"silent"))
print("At 'silent': ack actually true?", M.holds(("atom","ack"),"silent"))
print("\n=> obligation holds, but is violated. SDL distinguishes 'should' from 'is'.")
```

## 12. Quantum logic — non-distributive lattice

Propositions correspond to closed subspaces of a Hilbert space. The lattice of subspaces is **not distributive**: `a ∧ (b ∨ c) ≠ (a ∧ b) ∨ (a ∧ c)`. We sketch this with subspaces of `R²`.

```python
# Subspace = either {0}, a 1-D line through origin (represented by angle), or whole R^2.
ZERO = ("0",)
FULL = ("R2",)
def line(theta): return ("L", round(theta % math.pi, 9))

def meet(a,b):
    if a==FULL: return b
    if b==FULL: return a
    if a==ZERO or b==ZERO: return ZERO
    return a if a==b else ZERO
def join(a,b):
    if a==ZERO: return b
    if b==ZERO: return a
    if a==FULL or b==FULL: return FULL
    return a if a==b else FULL

a = line(0)            # x-axis
b = line(math.pi/4)    # 45°
c = line(math.pi/2)    # y-axis
lhs = meet(a, join(b,c))
rhs = join(meet(a,b), meet(a,c))
print("a ∧ (b ∨ c) =", lhs)
print("(a ∧ b) ∨ (a ∧ c) =", rhs)
print("Distributivity holds?", lhs == rhs)
print("\nAgents whose observations are mutually incompatible (think non-commuting measurements)")
print("need this kind of lattice instead of Boolean algebra.")
```

## 13. Linear logic — multiplicative resource accounting

Premises are *consumed* when used. Tensor `⊗` = both, lollipop `⊸` = consume left, produce right. Perfect for protocols where messages are physical resources (tokens, payments).

```python
from collections import Counter

def apply_rule(state: Counter, rule: Tuple[Counter, Counter]):
    """Rule = (consume, produce). Fires only if `consume` is available."""
    consume, produce = rule
    if all(state[k] >= v for k,v in consume.items()):
        new = Counter(state)
        new.subtract(consume); new += produce
        new = +new   # drop zero/negatives
        return new
    return None

# Agent A has 2 coins. Vending rule: 1 coin ⊸ 1 snack. Trade rule: 1 snack ⊸ 1 thanks.
state = Counter({"coin":2})
vend  = (Counter({"coin":1}), Counter({"snack":1}))
thank = (Counter({"snack":1}), Counter({"thanks":1}))
for step in range(4):
    print(f"step {step}: {dict(state)}")
    nxt = apply_rule(state, vend) or apply_rule(state, thank)
    if nxt is None: break
    state = nxt
print("\nClassical logic would let A 'have' a coin forever after spending it. Linear logic doesn't.")
```

## 14. Independence-friendly (IF) logic

Generalises FOL by letting a quantifier `∃y/∀x` mean "choose y *without knowing* x". Game-theoretic semantics: verifier vs falsifier with imperfect information. Models agents whose strategies cannot depend on hidden parameters.

We illustrate the schoolbook example: `∀x ∀z ∃y/∀z ∃w/∀x  (matrix)` — y must be picked without seeing z, w without seeing x.

```python
# Tiny domain {0,1}. Matrix: y == x  AND  w == z
domain = [0,1]

def verifier_wins_FOL():
    """Classical: ∀x∀z ∃y∃w (y==x ∧ w==z). Trivially true."""
    return all(any(any(y==x and w==z for w in domain) for y in domain) for x in domain for z in domain)

def verifier_wins_IF():
    """IF: y is a function of x only, w is a function of z only (independence).
    Verifier needs single y(x) and w(z) that win for all (x,z)."""
    for ystrat in product(domain, repeat=len(domain)):    # y(0), y(1)
        for wstrat in product(domain, repeat=len(domain)):# w(0), w(1)
            if all(ystrat[x]==x and wstrat[z]==z for x in domain for z in domain):
                return True
    return False

print("Verifier wins (classical FOL):", verifier_wins_FOL())
print("Verifier wins (IF, with independence):", verifier_wins_IF())
print("\nThey happen to coincide here, but IF logic captures cases where the gap matters —")
print("e.g., agents bidding without seeing each other's private signals.")
```

## Summary

| Logic | Communication scenario it nails |
|---|---|
| CTL | branching protocols, liveness under non-determinism |
| ATL | which coalitions can enforce what |
| DEL | private vs public observation, action models |
| Common knowledge | reasoning about iterated mutual knowledge |
| AGM revision | gracefully repairing beliefs after surprise |
| Possibilistic | weighted reliability of sources |
| Subjective logic | opinions with explicit uncertainty |
| Dempster–Shafer | combining evidence over hypothesis subsets |
| Bilattice (Belnap) | separate truth ordering from information ordering |
| Free logic | hypothetical or non-existent referents |
| SDL (deontic) | norms, obligations, permissions in protocols |
| Quantum | observations that are mutually incompatible |
| Linear | resource-consuming exchanges (tokens, payments) |
| IF logic | strategies under imperfect information |

**Cross-cutting experiment ideas:**
- Possibilistic + DEL: agents announce facts of varying necessity, common knowledge degrades gracefully.
- ATL + linear logic: which coalitions can guarantee an outcome *given finite resources*.
- Subjective logic + AGM: revise the *base rate* not the belief mass.
- Quantum + epistemic: agents whose knowledge is non-distributive — what does common knowledge even mean?
